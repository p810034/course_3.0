# Logistic Regression: Classification of Handwritten Digits

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

%matplotlib inline
digits = load_digits()
```

```python
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')

    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```

# Logistic Regression

# Train Test Data Split

```python
from sklearn.model_selection import train_test_split

# split the data into training and validation sets
X_train, X_test, y_train, y_test = train_test_split(digits.data,
                                                    digits.target,
                                                    test_size=0.25,
                                                    random_state=42)
```

# Build Logistic Regression Model & Train 

```python
from sklearn.linear_model import LogisticRegression
''' build your LR model here and please specify the value of C and\
try 'multinomial' classifier with 'sag' solver '''

# train the model
clf = LogisticRegression(C=1e-2, multi_class="ovr", solver='liblinear')

# fit on the training dataset
clf.fit(X_train, y_train)

# use the model to predict the labels of the test data
predicted = clf.predict(X_test)
expected = y_test
```

# Make prediction on Test data

```python
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(X_test.reshape(-1, 8, 8)[i],
              cmap=plt.cm.binary,
              interpolation='nearest')

    # label the image with the target value
    if predicted[i] == expected[i]:
        ax.text(0, 7, str(predicted[i]), color='green')
    else:
        ax.text(0, 7, str(predicted[i]), color='red')
```

這段代碼是用來可視化測試集中的數字圖像以及對應的預測標籤和真實標籤的。具體來說，它創建了一個 $8 \times 8$ 的子圖網格，每個子圖顯示一個數字圖像。如果預測標籤和真實標籤相同，則在圖像上方標註預測標籤，並使用綠色文本顏色，否則使用紅色文本顏色。

這段代碼使用了 Matplotlib 的 imshow() 函数將數字圖像顯示為灰度圖像。圖像數據被保存在 X_test 變量中，這是一個形狀為 (450, 64) 的 numpy 數組，其中每行代表一個 $8 \times 8$ 的數字圖像展平為一個 $64$ 維特徵向量。這些數字圖像來自於手寫數字資料集，稱為 digits，該資料集包含 $1797$ 張 $8 \times 8$ 的手寫數字圖像，分為 $10$ 個類別，每個類別包含 $0$ 到 $9$ 中的一個數字。predicted 和 expected 分別是用我們構建的 LogisticRegression 模型對測試集進行預測的結果和真實標籤。

# Quantitative Measurement on the Performance
印出accuracy, confusion matrix, f1-score, precision, recall等

```python
matches = (predicted == expected)
print('accuracy:%.3f' % (matches.sum() / float(len(matches))))
```

metrics.classification_report可以直接算出precision, recall, f1-score

```python
from sklearn import metrics
print(metrics.classification_report(expected, predicted))
```

# Confusion Matrix 

```python
print(metrics.confusion_matrix(expected, predicted))
```
